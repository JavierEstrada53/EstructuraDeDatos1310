```python
class NodoDoble:
    def __init__( self , value , siguiente = None , previo = None ):
        self.data = value
        self.next = siguiente
        self.prev = previo
```

```python
class DoubleLinkedList:
    def __init__( self  ):
        self.head = None
        self.tail = None
        self.size = 0

    def get_size( self ):
        if self.size == 0:
            print ('La lista est√° vac√≠a y por lo tanto no tiene tama√±o')
        else:
            print( f'La lista tiene {self.size} elementos')

    def get_size2( self ):
        return self.size
    
    def is_empty( self ): # true si la lista est√° vac√≠a
        return self.size == 0

    def append( self , value ):
        if self.is_empty():
            self.head = NodoDoble( value)
            self.tail = self.head
        else:
            self.tail.next = NodoDoble(value , None , self.tail )
            self.tail = self.tail.next
        self.size += 1

    def find_from_tail( self , value ):# --> buscar a partir de tail y regresar la posici√≥n
        curr_node = self.tail
        while curr_node.data != value:
            curr_node = curr_node.prev  
        if curr_node.data == value:
            print('El valor %s se encontr√≥ y est√° en la posicion %s' %(value,value))
        

    def find_from_head( self , value ):
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next
        if curr_node.data == value:
            print('El valor %s se encontr√≥ y est√° en la posicion %s' %(value,value))

    def remove_from_tail( self , value  ): # --> elimina a partir de tail
        curr_node = self.tail
        while curr_node.data != value:
            curr_node = curr_node.prev

        if curr_node.data == value:
            curr_node.next.prev = curr_node.prev
            curr_node.prev.next = curr_node.next
        self.size -= 1

    def remove_from_head( self , value  ): # --> elimina a partir de head
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next

        if curr_node.data == value:
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
        self.size -= 1
    
    def transversal( self ):
        curr_node = self.head
        while curr_node != None :
            print(f"| {curr_node.data} | --> ", end="")
            curr_node = curr_node.next
        print("")

    def reverse_transversal( self ):
        curr_node = self.tail
        while curr_node != None:
            print(f"| {curr_node.data} | --> ", end="")
            curr_node = curr_node.prev
        print("")
```

```python
#Probando los m√©todos
ldl = DoubleLinkedList()

#Get Size
print(f'Probando Get Size')
print(ldl.get_size())

#Is Empty
ldl.is_empty()

#Append
ldl.append(1)
ldl.append(2)
ldl.append(3)
ldl.append(4)
ldl.append(5)
ldl.append(6)



#Find from tail
print(f'\nProbando find from tail')
ldl.find_from_tail(4)

#Find from head
print(f'\nProbando find from head')
ldl.find_from_head(3)



print('\n')
ldl.transversal()

#Remove from tail
ldl.remove_from_tail (4)
print(f'\nDespu√©s del remove from tail con 40')
ldl.transversal()

#Remove from head
ldl.remove_from_head(3)
print(f'\nDespu√©s del remove from head con 30')
ldl.transversal()

#Transversal
print(f'\nProbando Transversal')
ldl.transversal()

#Reverse Transversal
print(f'\nProbando Reverse Transversal')
ldl.reverse_transversal()
```
