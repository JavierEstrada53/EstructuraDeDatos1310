```python
class Nodo:
    def __init__( self, value, siguiente = None ):
        self.data = value
        self.next = siguiente
```

```python
class LinkedList:
    def __init__( self ):
        self.head = None

    def is_empty( self ):
        return self.head == None
        
    def get_tail( self ):
        if self.is_empty():
            return None
        else:
            curr_node = self.head
            while curr_node.next != None:
                curr_node = curr_node.next
            return curr_node

    def append( self , value ):
        if self.head == None:
            self.head = Nodo ( value )
        else:
            curr_node = self.head
            while curr_node.next != None:
                curr_node = curr_node.next
            curr_node.next = Nodo( value )


    def transversal( self ):
        curr_node = self.head
        while curr_node != None:
            print(f'| {curr_node.data} | --> ', end='')
            curr_node = curr_node.next
        print('')

    def remove( self, value ):
        if self.head == None:
            print( 'La lista ligada est√° vac√≠a, no se puede aplicar "remove"')
        else:
            curr_node = self.head
            aux = None
            while curr_node.data != value: #revisar cuando el valor no exista
                aux = curr_node
                curr_node = curr_node.next
            if curr_node != None:
                aux.next = curr_node.next
                curr_node = None

    def preppend( self, value ):
        if self.head == None:
            print('La lista ligada est√° vac√≠a, no se puede aplicar "preppend"')
        else:
            aux = self.head
            self.head = Nodo ( value , aux )

#Implementaci√≥n de los m√©todos faltantes

    def pop( self ): #-->saca el √∫ltimo elemento
        if self.head == None:
            print('La lista ligada est√° vac√≠a, no se puede aplicar "pop"' )
        else:
            curr_node = self.head
            aux = None
            while curr_node.next != None: #  nos mueve al final
                aux = curr_node
                curr_node = curr_node.next
            if curr_node != None: # --> encontr√≥ el nodo final
                aux.next = curr_node.next
                curr_node = None

    def pop_first( self ):# --> saca el primer elemento
        if self.head == None:
            print('La lista ligada est√° vac√≠a, no se puede aplicar "pop first"' )
        else:
            curr_node = self.head
            aux = curr_node
            if curr_node != None: #estamos en head
                self.head = curr_node.next
```

```python
# Probando los m√©todos
lista = LinkedList()
if lista.is_empty() == True:
    print('La lista est√° vac√≠a')
else:
    print('La lista no est√° vac√≠a')

#append
lista.append(10)
lista.append(20)
lista.append(30)
lista.append(40)
lista.append(50)

#transversal
lista.transversal()

#get tail
nodo = lista.get_tail() #imprime el √∫ltimo valor
print(f'La cola es : {nodo.data}')

#preppend
lista.preppend(5)# agrega un 5 al inicio
print('Despu√©s del preppend con 5')
lista.transversal()

#remove
lista.remove(30)
print('Despu√©s del remove con 30')
lista.transversal()

#pop
lista.pop()
print('Despu√©s del pop con el √∫ltimo elemento')
lista.transversal()

#pop first
lista.pop_first()
print('Despu√©s del pop first con el 5')
lista.transversal()
```
