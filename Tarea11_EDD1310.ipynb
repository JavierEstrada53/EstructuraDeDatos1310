```python
class NodoBinario:
    def __init__( self , value , left = None , right = None ):
        self.data = value
        self.left = left
        self.right = right

class NodoBinarioArbol:
    def __init__( self ):
        self.root = None

    def insert( self , value ):
        if self.root == None:
            self.root = NodoBinario( value )
        else:
            self.__insert_nodo( self.root , value )     

    def __insert_nodo( self , nodo , value ):
        if nodo.data == value:
            pass
        elif value < nodo.data:
            if nodo.left == None:
                nodo.left = NodoBinario( value )
            else:
                self.__insert_nodo( nodo.left , value )
        else:
            if nodo.right == None:
                nodo.right = NodoBinario( value )
            else:
                self.__insert_nodo( nodo.right , value )

    def transversal( self , formato = "Inorden"):
        if formato == "Postorden":
            self.recorrido_pos( self.root )
        elif formato == "Preorden":
            self.recorrido_pre( self.root )
        else: # inorden
            self.recorrido_in( self.root )
        print()                         

    def recorrido_pos( self , nodo ):
        if nodo != None:
            self.recorrido_pos( nodo.left )
            self.recorrido_pos( nodo.right )
            print( str(nodo.data) , end = " , " )                       

    def recorrido_pre( self , nodo ):
        if nodo != None:
            print( str(nodo.data) , end = " , " )
            self.recorrido_pre( nodo.left )
            self.recorrido_pre( nodo.right )
                                   
    def recorrido_in( self , nodo ):
        if nodo != None:
            self.recorrido_in( nodo.left )
            print( str(nodo.data) , end = " , " )
            self.recorrido_in( nodo.right )

    def search( self , value ):
        if self.root == None:
            print("El árbol está vacío")
            return None
        else:
            self.__search( self.root , value )

    def __search( self , nodo , value ):
        if nodo.data == None:
            print("El dato no existe")
            return None

        elif nodo.data == value:
            print("Encontrado")
            return nodo

        elif value < nodo.data:
            return self.__buscar( nodo.left , value )

        else:
            return self.__buscar( nodo.right , value )

    def search_remove( self , nodo , value ): #método para encontrar el nodo
        if nodo == None:
            return None
        elif nodo.data == value:
            return nodo
        elif value < nodo.data:
            return self.search_remove( nodo.left , value )
        else:
            return self.search_remove( nodo.right , value )


    def remove( self , value ):
        if self.root == None:
            print(' El árbol está vacío ')
            return None
        else:
            encontrado = self.search_remove( self.root , value )
            if encontrado.left == None and encontrado.right == None: #caso 1, el nodo es hoja
                encontrado.data = None
            elif encontrado.left == None or encontrado.right == None: #caso 2, tiene 1 solo hijo
                if encontrado.left != None:
                    guardado = int( encontrado.left.data )
                    encontrado.data = None
                    encontrado = NodoBinario( guardado )
                if encontrado.right != None:
                    guardado = int( encontrado.right.data )
                    encontrado.data = None
                    encontrado = NodoBinario( guardado )
        
        
```
